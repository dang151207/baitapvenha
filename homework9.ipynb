## Bài 16
#### Cho hai số nguyên dương, n và m (0 < n, m <= 20 ). .Bạn hãy in ra tất cả các chuỗi nhị phân có n số 0 và m số 1. In theo thứ tự tăng dần trong từ điển 

```python
n, m = map(int,input().split())
def generate(s, zero, one):
    if zero == n and one == m:
        print(s)
        return
    if zero < n:
        generate(s + "0", zero + 1, one)
    if one < m:
        generate(s + "1", zero, one + 1)
generate("", 0, 0)
```

## Bài 17
#### Viết hàm tìm dãy con tăng nghiêm ngặt dài nhất trong dãy đầu vào từ bàn phím 

```python
def lis(n):
    ds = []
    for i in range(len(n)):
        if not ds or n[i] > ds[-1]:
            ds.append(n[i])
        else:
            pos = 0
            while pos < len(ds):
                if ds[pos] >= n[i]:
                    ds[pos] = n[i]
                    break
                pos += 1
    return ds
n = list(map(int,input().split()))
print(lis(n))

    
```

## Bài 18
#### Cho dãy N phần tử số nguyên, hãy đếm số cặp có vị trí khác nhau và có tổng bằng k (số nguyên dương cho vào từ bàn phím)

```python
n = list(map(int,input().split()))
k = int(input())
dem = 0 
for i in range(len(n)):
    for j in range(i+1, len(n)):
        if n[i] + n[j] == k:
            dem += 1
print(dem)
```

## Bài 19
#### Loại bỏ khỏi danh sách các phần tử trùng lặp, giữ lại đúng 1 lần xuất hiện, in lại theo đúng thứ tự xuất hiện

```python
n = list(map(int,input().split()))
ds = []
for i in n:
    if i not in ds:
        ds.append(i)
print(ds)
```
