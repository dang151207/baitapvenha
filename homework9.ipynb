## Bài 16
#### Cho hai số nguyên dương, n và m (0 < n, m <= 20 ). .Bạn hãy in ra tất cả các chuỗi nhị phân có n số 0 và m số 1. In theo thứ tự tăng dần trong từ điển 

```python
n, m = map(int,input().split())
def generate(s, zero, one):
    if zero == n and one == m:
        print(s)
        return
    if zero < n:
        generate(s + "0", zero + 1, one)
    if one < m:
        generate(s + "1", zero, one + 1)
generate("", 0, 0)
```

## Bài 17
#### Viết hàm tìm dãy con tăng nghiêm ngặt dài nhất trong dãy đầu vào từ bàn phím 

```python
def lis(n):
    ds = []
    for i in range(len(n)):
        if not ds or n[i] > ds[-1]:
            ds.append(n[i])
        else:
            pos = 0
            while pos < len(ds):
                if ds[pos] >= n[i]:
                    ds[pos] = n[i]
                    break
                pos += 1
    return ds
n = list(map(int,input().split()))
print(lis(n))

    
```

## Bài 18
#### Cho dãy N phần tử số nguyên, hãy đếm số cặp có vị trí khác nhau và có tổng bằng k (số nguyên dương cho vào từ bàn phím)

```python
n = list(map(int,input().split()))
k = int(input())
dem = 0 
for i in range(len(n)):
    for j in range(i+1, len(n)):
        if n[i] + n[j] == k:
            dem += 1
print(dem)
```

## Bài 20
Thực hiện chuẩn hóa chuỗi kí tự: 
a. bỏ khoảng trắng đầu cuối 
b. các khoảng trắng giữa chỉ còn 1 dấu cách 
c. viết hoa chữ cái đầu mỗi từ 
KHÔNG DÙNG CÁC HÀM SẴN CỐ

```python
s = input()

# Bỏ khoảng trắng thừa
res = ""
space = True

for c in s:
    if c == " ":
        if not space:
            res += c
            space = True
    else:
        res += c
        space = False

# bỏ khoảng trắng cuối
while len(res) > 0 and res[-1] == " ":
    res = res[:-1]

# bỏ khoảng trắng đầu
while len(res) > 0 and res[0] == " ":
    res = res[1:]

# viết hoa đầu từ
ans = ""
new_word = True

for c in res:
    if c == " ":
        ans += c
        new_word = True
    else:
        if new_word and 'a' <= c <= 'z':
            ans += chr(ord(c) - 32)
        else:
            ans += c
        new_word = False

print(ans)
```

## Bài 21
#### Tìm phần tử lớn thứ k trong dãy n phần tử bằng thuật toán tốt nhất có thể – Tránh dùng sắp xếp (dùng phân hoạch của quicksort) 

```python
n = int(input())
a = list(map(int, input().split()))
k = int(input())

top = []

for x in a:
    if x not in top:
        top.append(x)
        top.sort(reverse=True)

        if len(top) > k:
            top.pop()

if len(top) < k:
    print("Khong ton tai")
else:
    print(top[-1])
```

## Bài 22

```python
s1 = input()
s2 = input()

if len(s1) != len(s2):
    print("NO")
else:
    cnt1 = {}
    cnt2 = {}

    for c in s1:
        cnt1[c] = cnt1.get(c, 0) + 1

    for c in s2:
        cnt2[c] = cnt2.get(c, 0) + 1

    if cnt1 == cnt2:
        print("YES")
    else:
        print("NO")
```

## Bài 23
#### Cho số nguyên rất lớn (string), và một số nguyên dương k. Tìm số có k chữ số lớn nhất có thể từ số ban đầu bằng cách xóa bớt đi một số chữ số nào đó mà không thay đổi vị trí của bất kì chữ số nào. (VD: 78934340999, k = 7 => 9440999 là số lớn nhất có 7 chữ số cần tìm) 

```python
s = input()
k = int(input())

max_num = ""

for i in range(len(s) - k + 1):
    sub = s[i:i+k]

    if sub > max_num:
        max_num = sub

print(max_num)
```

## Bài 24
#### Nén chuối kí tự bằng cách đếm số lần xuất hiện của các kí tự liên tiếp nhau và ghi lại ngay sau kí tự đó. VD aaaaaafdhhhhh =>a6f1d1h5

```python
s = input()

cnt = 1
res = ""

for i in range(1, len(s)):
    if s[i] == s[i-1]:
        cnt += 1
    else:
        res += s[i-1] + str(cnt)
        cnt = 1

res += s[-1] + str(cnt)

print(res)
```
