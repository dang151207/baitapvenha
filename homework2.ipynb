```python
a,b,c= map(int,input().split())
if a+b>c and a+c>b and b+c>a:
    def kiem_tra_tam_giac(a,b,c):
        if a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
            return "đây là tam giác vuông"
        elif a==b==c:
            return "đây là tam giác đều"
        elif a==b or a==c or b==c:
            return "đây là tam giác cân"
        else:
            return "đây là tam giác thường"
    print(kiem_tra_tam_giac(a,b,c))
    p= (a+b+c)/2
    dien_tich_tg= (p*(p-a)*(p-b)*(p-c))**0.5
    chu_vi_tg= 2*p
    print("diện tích tam giác là", f"{dien_tich_tg:.2f}")
    print("chu vi tam giác là", f"{chu_vi_tg:.2f}")
else:
    print("Khong la ba canh cua mot tam giac")
```

## Bài 5


```python
import math
a,b,c= map(float,input().split())
if a==0:
    if b==0:
        if c==0:
            print("phương trình có vô số nghiệm")
        else:
            print("phương trình vô nghiệm")
    else:
        x= -c/b
        print(f"nghiệm là x = {x}")
else:
    delta = b**2 - 4*a*c
    if delta > 0 :
        x1= round((-b- math.sqrt(delta))/(2*a),2)
        x2= round((-b+ math.sqrt(delta))/(2*a),2)
        print(f"nghiệm là x1 = {x1} và x2 = {x2}")
    elif delta==0:
        x= -b/(2*a)
        print(f"nghiệm kép là x = {x}")
    else:
        print("phương trình vô nghiệm")
```

```python
a, b, c, d, m, n= map(float,input().split())
D1 = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c
if D1 != 0:
    x = Dx / D1
    y = Dy / D1
    print("Nghiệm:")
    print("x =", x)
    print("y =", y)
else:
    if Dx == 0 and Dy == 0:
        print("Hệ có vô số nghiệm")
    else:
        print("Hệ vô nghiệm")
```
