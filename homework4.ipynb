## ex 2


```python
def uoc_so_nguyen_duong(n):
    dem = 0
    cac_uoc = []
    tong = 0
    for i in range(1, n+1):
        if n%i == 0:
            dem += 1
            cac_uoc.append(i)
            tong += i
    return f"các ước nguyên dương là {cac_uoc}, tổng các ước là {tong}, có {dem} ước"
n= int(input())           
print(uoc_so_nguyen_duong(n))   
```

```python
def tong_uoc_nguyen_duong(n):
    tong = 0
    for i in range(1,n):
        if n%i == 0:
            tong += i
    return tong == n
n= int(input())
perfect_num = []
for i in range(1,n+1):
    if tong_uoc_nguyen_duong(i):
        perfect_num.append({i})
print("các số hoàn hảo(perfect number) trong khoảng (1,n) là", perfect_num)
```
