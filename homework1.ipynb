## Bài 8

```python
def tinh_tien_dien(so_dien):
    if so_dien <= 50 :
        tien = so_dien*1984
    elif 50< so_dien <= 100 :
        tien = 50*1984 + (so_dien-50)*2050
    elif 100< so_dien <= 200 :
        tien = 50*1984 + 50*2050 + (so_dien-100)*2380
    elif 200< so_dien <= 300 :
        tien = 50*1984 + 50*2050 + 100*2380 +(so_dien-200)*2998
    elif 300< so_dien <= 400 :
        tien = 50*1984 + 50*2050 + 100*2380 + 100*2998 +(so_dien-300)*3350
    else:
        tien = 50*1984 + 50*2050 + 100*2380 + 100*2998 + 100*3350 +(so_dien-400)*3460
    return tien
ten_chu_ho = input()
chi_so_thang_truoc= int(input())
chi_so_thang_nay= int(input())
so_dien = chi_so_thang_nay - chi_so_thang_truoc
tien = tinh_tien_dien(so_dien)
tien_vat= int(tien * 1.08)
print(f"Họ và tên: {ten_chu_ho}")
print(f"Tiền phải trả là: {tien_vat}")
```
