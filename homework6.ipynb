## Bài 20


```python
def tim_k(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

a = list(map(int, input().split()))
k = int(input())

print(tim_k(a, k))
```

## Bài 21

```python
s = input().split()

d = {}

for i in range(len(s)):
    if s[i] not in d:
        d[s[i]] = []
    d[s[i]].append(i)

for key in d:
    d[key] = tuple(d[key])

print(d)
```

## Bài 24


```python
def day_0_ve_cuoi(lst):
    kq = []

    for x in lst:
        if x != 0:
            kq.append(x)

    so0 = lst.count(0)

    for i in range(so0):
        kq.append(0)

    return kq

a = list(map(int, input().split()))
print(day_0_ve_cuoi(a))
```

```python
def day_0_ve_dau(lst):
    so0 = lst.count(0)

    kq = [0] * so0

    for x in lst:
        if x != 0:
            kq.append(x)

    return kq

a = list(map(int, input().split()))
print(day_0_ve_dau(a))
```

## Bài 26


```python
a = list(map(int, input().split()))

tong = 0

for i in range(1, len(a) - 1):
    if a[i] > a[i - 1] and a[i] > a[i + 1]:
        tong += a[i]

print(tong)
```

## Bài 27


```python
a = list(map(int, input().split()))
x = int(input())

vt = []

for i in range(len(a)):
    if a[i] < x:
        vt.append(i)

if len(vt) == 0:
    print(-1)
else:
    print(*vt)
```

## Bài 28


```python
a = list(map(int, input().split()))

tang = True
giam = True

for i in range(len(a) - 1):
    if a[i] >= a[i + 1]:
        tang = False

    if a[i] <= a[i + 1]:
        giam = False

if tang:
    print("Tang thuc su")
elif giam:
    print("Giam thuc su")
else:
    print("Vo danh")
```
