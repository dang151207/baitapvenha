```python
a= int(input())
def so_chinh_phuong(a):
    for i in range(1, int(a**0.5)+1):
        if i**2==a:
            return f"{a} là số chính phương"
    return f"{a} ko phải số chính phương"
print(so_chinh_phuong(a))

```

## Bài  10

```python
def so_hoan_hao(n):
    tong=0
    for i in range(1,n):
        if n%i == 0:
            tong += i
    if tong==n:
        return f"{n} là số hoàn hảo"
    else:
        return f"{n} ko phải số hoàn hảo"

n= int(input())
print(so_hoan_hao(n))
```

## Bài 20

```python
# cách 1
def so_nguyen_to(n):
    if n<2:
        return False
    else:
        for i in range(2, int(n**0.5)+1):
            if n % i == 0:
                return False
        return True
n= int(input())
m=n
dem= 0
while n>0:
    k= n%10
    if so_nguyen_to(k):
        dem += 1
    n //= 10
print(f"số {m} có {dem} chữ số nguyên tố")
    
```

```python
# cách 2
n = int(input())
m = n
dem = 0

while n > 0:
    k = n % 10
    
    if k in [2, 3, 5, 7]:
        dem += 1
    
    n //= 10

print(f"số {m} có {dem} chữ số nguyên tố")
```

```python
# cách 3: trả về số chữ số nguyên tố ko trùng nhau
n = int(input())
m = n
digits = []

while n > 0:
    k = n % 10
    
    if k in [2, 3, 5, 7] and k not in digits:
        digits.append(k)
    
    n //= 10

print(f"số {m} có {len(digits)} chữ số nguyên tố khác nhau")
```
