## Bài 16
#### Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử x trong mảng xuất hiện đúng x lần. Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.


```python
n = list(map(int,input().split()))
ds = {}
for i in n:
    if i not in ds:
        ds[i] = 1
    else:
        ds[i] += 1
y = 0
for x in ds:
    if ds[x] > x:
        y += ds[x] - x
    elif ds[x] < x:
        y += ds[x]

if y == 0:
    print("Mang dep")
else:
    print(f"Mang ko dep, can xoa {y} so")
```

```python
def mang_dep(n):
    dem = {}
    for i in n:
        if i not in dem:
            dem[i] = 1
        else:
            dem[i] += 1

    y = 0
    for x in dem:
        # trường hợp số lần xuất hiện lớn hơn số trong mảng nên thêm y số
        if dem[x] > x:
           y += dem[x] - x
        # trường hợp số lần xuất hiện bé hơn số trong mảng nên cần xóa luôn số đó
        elif dem[x] < x: 
           y += dem[x]
    if y == 0:
        return "Mang dep"
    else:
        return f"Mang ko dep, can xoa {y} so"
lst = list(map(int,input().split()))
print(mang_dep(lst))
```

## Bài 17
#### Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. Nếu không có phần tử nào thỏa mãn thì không in gì


```python
n = list(map(int,input().split()))
candidate = None 
dem = 0
for i in n:
    if dem == 0:
        candidate = i
        dem = 1
    else:
        if i == candidate:
            dem += 1
        else:
            dem -= 1
if n.count(candidate) > len(n) // 2:
    print(candidate)
```

```python
a = list(map(int,input().split()))
b = list(map(int,input().split()))
i = 0
j = 0
ds = []
tang = a[0] < a[1]
while i < len(a) and j < len(b):
    if tang:
        if a[i] < b[j]:
            ds.append(a[i])
            i += 1
        else:
            ds.append(b[j])
            j += 1
    else:
        if a[i] > b[j]:
            ds.append(a[i])
            i += 1
        else:
            ds.append(b[j])
            j += 1
while i < len(a):
    ds.append(i)
    i += 1
while j < len(b):
    ds.append(b[j])
    j += 1
print(ds)
```

```python
n = list(map(int,input().split()))
k = int(input())
tong = 0
for i in n:
    m = i
    dem = 0
    while i > 0:
        dem += 1
        i //= 10
    if dem == k:
        tong += m
print(tong)
        
```

## Bài 25
#### Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số

```python
n = input()
k = int(input())
ds = []
xoa  = len(n) - k
for i in n:
    while ds and xoa > 0 and ds[-1] < i:
        ds.pop()
        xoa -= 1
    ds.append(i)
while xoa > 0:
    ds.pop()
    xoa -= 1
print("".join(ds))
```

```python
n = input()
k = int(input())

xoa = len(n) - k # số kí tự phải xóa trong n
stack = []
for i in n:
    # nếu phần tử cuối của stack bé hơn i -> xóa phần tử cuối đó
    while stack and xoa > 0 and stack[-1] < i:
        stack.pop()
        xoa -= 1
    # thêm i vào stack nếu phần tử cuối của stack lớn hơn i
    stack.append(i)
# trường hợp xét hết rồi nhưng xóa vẫn bằng 0. ví dụ 9876
# xóa nốt các phần tử thừa trong stack
while xoa > 0:
    stack.pop()
    xoa -= 1
print("".join(stack))
```

```python
n = list(map(int,input().split()))
max = 0
dem = 1
start = 0
best_start = 0
for i in range(1,len(n)):
    if n[i] > n[i-1]:
        dem += 1
    else:
        if dem > max:
            max = dem
            best_start = start
        dem = 1
        start = i
if dem > max:
    max = dem
    best_start = start
ket_qua = n[best_start : best_start + max]
print(ket_qua)
```

## Bài 30
#### Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím

```python
n = list(map(int,input().split()))
ds = []
for i in range(len(n)):
    if not ds or n[i] > ds[-1]:
        ds.append(n[i])
    else:
        for x in range(len(ds)):
            if ds[x] >= n[i]:
                ds[x] = n[i]
                break
print(ds)
                
```
