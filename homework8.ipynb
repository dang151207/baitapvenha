Xây dựng lớp Shape với phương thức area().
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area().
Lấy π = 3.14.
Input
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE)
Dòng 2: tham số tương ứng.
Output
In ra diện tích với 2 chữ số thập phân. 


```python
class shape:
    def area(self):
        return 0
class rectangle(shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h
class circle(shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return ((self.r)**2) * 3.14

shape_type = input()
if shape_type == "RECTANGLE":
    w, h = map(float,input().split())
    shape = rectangle(w, h)
else:
    r = float(input())
    shape = circle(r)

print(f"{shape.area():.2f}")
```

Xây dựng lớp Polygon với phương thức perimeter().
Xây dựng lớp Triangle(a, b, c) kế thừa từ Polygon và ghi đè perimeter() trả về a+b+c.
Input
Một dòng chứa 3 số nguyên a b c.
Output
In ra chu vi tam giác

```python
class polygon:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self):
        return self.a + self.b + self.c
class triangle(polygon):
    def __init__(self,a,b,c):
        super().__init__(a,b,c)
    def perimeter(self):
        return self.a + self.b + self.c

a,b,c = map(int,input().split())
print(triangle(a,b,c).perimeter())
```

Xây dựng lớp Account(balance) với hai phương thức deposit(x) và get_balance(). 
Xây dựng lớp SavingAccount(balance, rate) kế thừa từ Account, thêm phương thức 
add_interest() để cộng balance*rate vào số dư. 
Input 
balance rate 
Output 
Số dư sau khi cộng lãi, in 2 chữ số thập phân.

```python
class Account:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, x):
        self.balance = self.balance + x
        return self.balance
    def get_balance(self):
        return self.balance
class SavingAccount(Account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = rate
    def add_interest(self):
        self.balance = self.balance + self.balance*self.rate
        return self.balance
balance, rate = map(float,input().split())
result = SavingAccount(balance, rate)
print(f"{result.add_interest():.2f}")
```

Xây dựng lớp Printer với phương thức print_value(x). 
Xây dựng lớp BinaryPrinter kế thừa từ Printer, ghi đè print_value(x) để in biểu diễn nhị phân của 
số nguyên dương x. 
Input 
Một số nguyên dương n. 
Output 
In ra dạng nhị phân của n.

```python
class Printer:
    def print_value(self, x):
        print(x)
class BinaryPrinter(Printer):
    def print_value(self, x):
        binary = ""
        while x > 0:
            binary = str(x%2) + binary
            x //= 2
        print(binary)
x = int(input())
result = BinaryPrinter()
result.print_value(x)
```

Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận: 
[a b] 
[c d] 
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc. 
Input 
Một dòng chứa 4 số nguyên a b c d. 
Output 
In ra trace và determinant trên cùng một dòng.

```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        return self.a + self.d
    def determinant(self):
        return self.a*self.d - self.b*self.c
a,b,c,d = map(int,input().split())
result = Matrix2x2(a,b,c,d)
print(f"trace: {result.trace()}, determinant: {result.determinant()}")
```

Xây dựng lớp Product(name, price). 
Xây dựng lớp DiscountProduct(name, price, percent) kế thừa từ Product, ghi đè phương thức 
final_price() để trả về giá sau giảm. 
Công thức: price * (100 - percent) / 100 
Input 
name price percent
Output 
In ra giá cuối cùng với 2 chữ số thập phân.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = float(price)
    def final_price(self):
        print(self.price)
class DiscountProduct(Product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = float(percent)
    def final_price(self):
        print(f"{(self.price * (100 - self.percent) / 100):.2f}")
name, price, percent = input().split()
result = DiscountProduct(name, price, percent)
result.final_price()
```

Yêu cầu 
Xây dựng lớp Counter có thuộc tính count ban đầu bằng 0 và các phương thức inc(), dec(), 
value(). 
Sau đó thực hiện dãy thao tác trên đối tượng. 
Input 
Dòng 1: n 
Dòng 2: n thao tác, mỗi thao tác là INC hoặc DEC, cách nhau bởi dấu cách. 
Output 
Giá trị cuối cùng của count.

```python
class Counter:
    def __init__(self):
        self.count = 0
    def inc(self):
        self.count += 1
        return self.count
    def dec(self):
        self.count -= 1
        return self.count
    def value(self):
        return self.count

n = int(input())
thao_tac = input().split()
result = Counter()
for i in thao_tac:
    if i == "INC":
        result.inc()
    elif i == "DEC":
        result.dec()
print(result.value())
```

Xây dựng lớp Time(h, m, s) với phương thức to_seconds() đổi thời gian sang tổng số giây. 
Input 
h m s 
Output 
Tổng số giây.

```python
class Time:
    def __init__(self, h, m, s):
        self.h = h
        self.m = m
        self.s = s
    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s
h,m,s = map(int,input().split())
result = Time(h,m,s)
print(result.to_seconds())
```

Yêu cầu 
Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x), 
pop(), top(). 
Dữ liệu đảm bảo mọi lệnh pop và top đều hợp lệ. 
Input 
Dòng 1: n 
n dòng tiếp theo là một trong ba lệnh: 
PUSH x 
POP 
TOP 
Output 
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng.

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self,x):
        self.stack.append(x)
    def pop(self):
        self.stack.pop()
    def top(self):
        return self.stack[-1]
    
n = int(input())
result = Stack()
ds_top = []
for i in range(n):
    lenh = input().split()
    if lenh[0] == "PUSH":
        result.push(int(lenh[1]))
    elif lenh[0] == "POP":
        result.pop()
    elif lenh[0] == "TOP":
        ds_top.append(result.top())
print(*ds_top)
```
