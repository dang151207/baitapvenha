## Bài 16
#### Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x


```python
def armstrong(n):
    tong = 0
    m = n
    while n>0 :
        k = n % 10
        tong += k**3
        n //= 10
    if tong == m:
        return True
    return False

n = int(input())

print(armstrong(n))
```

## Bài 17
#### Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def so_chinh_phuong(n):
    for i in range(n+1):
        if i**2 == n:
            return True 
    return False

n= int(input())

print(so_chinh_phuong(n))
```

## Bài 18
#### VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def so_doi_xung(n):
    m = n
    so_moi = 0
    while n>0 :
        so_cuoi = n % 10
        so_moi = so_moi*10 + so_cuoi
        n //= 10
    if so_moi == m:
        return True
    return False

n = int(input())

print(so_doi_xung(n))
```

## Bài 19
#### VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def so_ki_dieu(n):
    ds = set()
    while n != 1:
        if n in ds:
            return False
        ds.add(n)
    
        tong = 0
        while n>0:
            k = n%10
            tong += k**2
            n //= 10
        n = tong
    return True
n= int(input())
print(so_ki_dieu(n))
```

## Bài 20
#### Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
def detect_number_type(s):
    if type(s) == int:
        return "int"
    elif type(s) == float:
        return "float"
    else:
        return "neither"
print(detect_number_type(1))
```

## Bài 21
#### Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số 

```python
def is_one_product_of_others(a,b,c):
    if a == b*c or c == b*a or b == c*a:
        return True
    return False
a,b,c= map(int,input().split())
print(is_one_product_of_others(a,b,c))
```

## Bài 23
#### Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a,b,c):
    if a == 0:
        if b == 0:
            if c == 0:
                return "phương trình vô số nghiệm"
            else:
                return "phương trình vô nghiệm"
        else:
            x = round(-c/b,2)
            return f"x={x}"
    else:
        delta = b**2 -4*a*c
        if delta >= 0:
            x1 = round((-b-delta**0.5)/(2*a),2)
            x2 = round((-b+delta**0.5)/(2*a),2)
            if x1 == x2:
                return x1
            else:
                return x1, x2
        else:
            return "phương trình vô nghiệm"
        
a,b,c = map(int,input().split())

print(quadratic_roots_count(a,b,c))
           
```
